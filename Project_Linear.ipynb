```python
import numpy as np
import scipy.linalg as la
```

```python
def diet_optimizer(A, b, ingredients, nutrients):
    print("=== Diet Optimizer ===")
    print(f"วัตถุดิบที่มี: {ingredients}")
    print(f"เป้าหมายโภชนาการ: {nutrients} = {b}\n")

    # ---------------------------------------------------------
    # 1. การตรวจสอบ Range (พิสัย) ของเมทริกซ์
    # ทฤษฎี: เช็คว่าเป้าหมาย (b) เป็นไปได้ไหม (อยู่ใน Range ของ A หรือเปล่า)
    # โดยเปรียบเทียบ Rank ของเมทริกซ์ A กับ Rank ของเมทริกซ์แต่งเติม [A|b]
    # ---------------------------------------------------------
    rank_A = np.linalg.matrix_rank(A)
    augmented_A = np.column_stack((A, b))
    rank_augmented = np.linalg.matrix_rank(augmented_A)

    print("--- การวิเคราะห์ Range และ Kernel ---")
    if rank_A < rank_augmented:
        print("เป้าหมายโภชนาการนี้ไม่อยู่ใน Range (ไม่สามารถจัดสัดส่วนให้ได้เป้าหมายเป๊ะๆ ได้)")
        return
    else:
        print("เป้าหมายโภชนาการนี้อยู่ใน Range (สามารถจัดสัดส่วนวัตถุดิบได้)")

    # ---------------------------------------------------------
    # 2. การตรวจสอบ Kernel (Null Space)
    # ทฤษฎี: ถ้า Kernel ไม่ใช่เซตว่าง (มีมิติ > 0) แปลว่ามีหลายสูตรที่ให้สารอาหารเท่ากันเป๊ะ
    # ---------------------------------------------------------
    null_space = la.null_space(A)
    if null_space.size > 0:
        print("เมทริกซ์มี Kernel (Nullity > 0) หมายความว่ามีสูตรอาหาร 'มากกว่า 1 วิธี' ที่ให้สารอาหารตามเป้า")
    else:
        print("เมทริกซ์ไม่มี Kernel (Nullity = 0) หมายความว่ามีสูตรอาหารนี้เพียง 'วิธีเดียวเท่านั้น' (Unique Solution)\n")

    # ---------------------------------------------------------
    # 3. การแก้สมการด้วย LU Decomposition (A = LU)
    # ทฤษฎี: แยกตัวประกอบเมทริกซ์ เพื่อให้การคำนวณเป้าหมายใหม่ (b) ในอนาคตทำได้รวดเร็ว
    # ---------------------------------------------------------
    print("--- การคำนวณสัดส่วนด้วย LU Decomposition ---")
    # หาค่า P, L, U
    P, L, U = la.lu(A)
    print("Matrix L (Lower Triangular):\n", np.round(L, 2))
    print("Matrix U (Upper Triangular):\n", np.round(U, 2), "\n")

    # ใช้ LU Decomposition ในการแก้สมการ Ax = b
    # Scipy มีฟังก์ชัน lu_factor และ lu_solve เพื่อแก้สมการจาก LU ได้ทันที
    lu_pivot = la.lu_factor(A)
    x = la.lu_solve(lu_pivot, b)

    print("--- สรุปสัดส่วนวัตถุดิบที่ต้องใช้ ---")
    for i in range(len(ingredients)):
        print(f"{ingredients[i]}: {x[i]:.2f} หน่วย (เช่น ขีด หรือ 100 กรัม)")
```

```python
# ==========================================
# ส่วนจำลองข้อมูล (สมมติว่า 1 หน่วย = 100 กรัม)
# ==========================================
# เมทริกซ์ A (คอลัมน์ = อกไก่, ข้าวกล้อง, บรอกโคลี)
# แถว 1: โปรตีน (g)
# แถว 2: คาร์โบไฮเดรต (g)
# แถว 3: แคลอรี (kcal)
A_matrix = np.array([
    [30, 3, 3],    # โปรตีนที่ได้จากวัตถุดิบทั้ง 3
    [0, 30, 7],    # คาร์บที่ได้จากวัตถุดิบทั้ง 3
    [165, 130, 35] # แคลอรีที่ได้จากวัตถุดิบทั้ง 3
])

# เวกเตอร์ b (เป้าหมายที่เราอยากได้)
# สมมติเราอยากกินให้ได้: โปรตีน 72g, คาร์บ 74g, แคลอรี 660 kcal
b_vector = np.array([72, 74, 660])

ingredients_list = ["อกไก่", "ข้าวกล้อง", "บรอกโคลี"]
nutrients_list = ["โปรตีน(g)", "คาร์บ(g)", "แคลอรี(kcal)"]
```

```python
diet_optimizer(A_matrix, b_vector, ingredients_list, nutrients_list)
```
